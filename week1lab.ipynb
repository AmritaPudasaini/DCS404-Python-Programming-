### 3.2: Exercise on List Manipulation

```python
# Extract every other element
def extract_every_other(lst):
    return lst[::2]

# Return sublist from indexes
def get_sublist(lst, start, end):
    return lst[start:end+1]

# Reverse list using slicing
def reverse_list(lst):
    return lst[::-1]

# Remove first and last
def remove_first_last(lst):
    return lst[1:-1]

# Get first n elements
def get_first_n(lst, n):
    return lst[:n]

# Get last n elements
def get_last_n(lst, n):
    return lst[-n:]

# Reverse skipping one element
def reverse_skip(lst):
    return lst[-2::-2]

# Example list
numbers = [1, 2, 3, 4, 5, 6]
print("Original List:", numbers)
print("Every Other Element:", extract_every_other(numbers))
print("Sublist (2 to 4):", get_sublist(numbers, 2, 4))
print("Reversed List:", reverse_list(numbers))
print("Without First and Last:", remove_first_last(numbers))
print("First 3 Elements:", get_first_n(numbers, 3))
print("Last 2 Elements:", get_last_n(numbers, 2))
print("Reverse Skip List:", reverse_skip(numbers))
```

### 3.3 Exercise on Nested List

#### Flatten a Nested List

```python
def flatten(lst):
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)

    return result


nested_list = [[1, 2], [3, 4], [5]]

print("Flattened List:", flatten(nested_list))
```

#### Accessing Nested List Elements

```python
def access_nested_element(lst, indices):
    element = lst

    for index in indices:
        element = element[index]

    return element

lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print("Element:", access_nested_element(lst, [1, 2]))
```

#### Sum of All Elements in a Nested List

```python
def sum_nested(lst):
    total = 0

    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item

    return total

nested_list = [[1, 2], [3, [4, 5]], 6]
print("Sum:", sum_nested(nested_list))
```

#### Remove Specific Element from Nested List

```python
def remove_element(lst, elem):
    result = []

    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)

    return result

lst = [[1, 2], [3, 2], [4, 5]]
print("Updated List:", remove_element(lst, 2))
```

#### Find Maximum Element in Nested List

```python
def flatten(lst):
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)

    return result

def find_max(lst):
    return max(flatten(lst))

nested_list = [[1, 2], [3, [4, 5]], 6]
print("Maximum Element:", find_max(nested_list))
```

#### Count Occurrences in Nested List

```python
def count_occurrences(lst, elem):
    count = 0

    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

lst = [[1, 2], [2, 3], [2, 4]]
print("Occurrences of 2:", count_occurrences(lst, 2))
```

#### Flatten List of Lists of Lists

```python
def deep_flatten(lst):
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

lst = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print("Deep Flattened List:", deep_flatten(lst))
```

#### Nested List Average

```python
# Flatten nested list
def flatten(lst):
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)

    return result


# Calculate average
def average_nested(lst):
    flat_list = flatten(lst)
    return sum(flat_list) / len(flat_list)

nested_list = [[1, 2], [3, 4], [5, 6]]
print("Average:", average_nested(nested_list))
```

## Problem- 4: Matrix Operations

#### 1. Prove A × A⁻¹ = I

```python
import numpy as np
A = np.array([[3, 4], [7, 8]])
A_inv = np.linalg.inv(A)
result = np.dot(A, A_inv)
print(result)
```

#### 2. Prove AB ≠ BA

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Multiply AB
AB = np.dot(A, B)

# Multiply BA
BA = np.dot(B, A)

print("AB =")
print(AB)

print("\nBA =")
print(BA)
```

#### 3. Prove (AB)ᵀ = BᵀAᵀ

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

left = np.dot(A, B).T
right = np.dot(B.T, A.T)

print("Left Side:")
print(left)

print("\nRight Side:")
print(right)
```

#### Solve Linear Equation Using Inverse Method

```python
import numpy as np

A = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

B = np.array([-1, -3, 9])

# Find inverse of A
A_inv = np.linalg.inv(A)

# Solve equation
X = np.dot(A_inv, B)

print("Solution:")
print(X)
```

## 9.2 Experiment: How Fast is Numpy?

### 1. Element-wise Addition

#### Python

```python
import time
n = 1_000_000
a = list(range(n))
b = list(range(n))

start = time.time()

result = []
for i in range(n):
    result.append(a[i] + b[i])

end = time.time()
print("List Addition Time:", end - start)
```

#### NumPy Array

```python
import numpy as np
import time

n = 1_000_000

a = np.arange(n)
b = np.arange(n)

start = time.time()

result = a + b
end = time.time()
print("NumPy Addition Time:", end - start)
```

#### 4. Matrix Multiplication

#### Python List

```python
import time
size = 1000

A = [[1 for _ in range(size)] for _ in range(size)]
B = [[1 for _ in range(size)] for _ in range(size)]

start = time.time()
result = []
for i in range(size):
    row = []
    for j in range(size):
        s = 0
        for k in range(size):
            s += A[i][k] * B[k][j]
        row.append(s)
    result.append(row)
end = time.time()
print("List Matrix Multiplication Time:", end - start)
```

#### NumPy List

```python
import numpy as np
import time

size = 1000

A = np.ones((size, size))
B = np.ones((size, size))

start = time.time()

result = np.dot(A, B)
end = time.time()
print("NumPy Matrix Multiplication Time:", end - start)
```
